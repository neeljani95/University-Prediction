```python
from tkinter import *
from tkinter import messagebox
import joblib
import pandas as pd

df2 = pd.read_csv("3 - database_2.csv (2) 2 - in (1).csv")
df2.columns = df2.columns.str.replace(' ','_')

dlink2 = df2.drop(['Links'], axis=1, inplace=True)
dlogo2 = df2.drop(['logo'], axis=1, inplace=True)

df = pd.read_csv("3 - database_2.csv (2) 2 - in (1).csv")
df.columns = df.columns.str.replace(' ','_')

dlink = df.drop(['Links'], axis=1, inplace=True)
dlogo = df.drop(['logo'], axis=1, inplace=True)



t1 = tuple(df['country'])
t2 = tuple(df['city'])
t3 = tuple(df['region'])
t4 = tuple(df['type'])
t5 = tuple(df['size'])
t6 = tuple(df['sports'])
t7 = tuple(df['percentage'])
t8 = tuple(df['Music'])
t9 = tuple(df['University'])
t10 = tuple(df['University_Names'])

factorized_categories_1 = pd.Series(t1)
factorized_categories_2 = pd.Series(t2)
factorized_categories_3 = pd.Series(t3)
factorized_categories_4 = pd.Series(t4)
factorized_categories_5 = pd.Series(t5)
factorized_categories_6 = pd.Series(t6)
factorized_categories_7 = pd.Series(t7)
factorized_categories_8 = pd.Series(t8)


unique_categories_1 = factorized_categories_1.unique()
unique_categories_2 = factorized_categories_2.unique()
unique_categories_3 = factorized_categories_3.unique()
unique_categories_4 = factorized_categories_4.unique()
unique_categories_5 = factorized_categories_5.unique()
unique_categories_6 = factorized_categories_6.unique()
unique_categories_7 = factorized_categories_7.unique()
unique_categories_8 = factorized_categories_8.unique()

category_dict_1 = {code: category for code, category in enumerate(unique_categories_1)}
category_dict_2 = {code: category for code, category in enumerate(unique_categories_2)}
category_dict_3 = {code: category for code, category in enumerate(unique_categories_3)}
category_dict_4 = {code: category for code, category in enumerate(unique_categories_4)}
category_dict_5 = {code: category for code, category in enumerate(unique_categories_5)}
category_dict_6 = {code: category for code, category in enumerate(unique_categories_6)}
category_dict_7 = {code: category for code, category in enumerate(unique_categories_7)}
category_dict_8 = {code: category for code, category in enumerate(unique_categories_8)}

df['country'] = df['country'].factorize()[0]
df['city'] = df['city'].factorize()[0]
df['type'] = df['type'].factorize()[0]
df['region'] = df['region'].factorize()[0]
df['size'] = df['size'].factorize()[0]
df['sports'] = df['sports'].factorize()[0]
df['percentage'] = df['percentage'].factorize()[0]
df['Music'] = df['Music'].factorize()[0]

university_dict = dict(zip(df['University'], df['University_Names']))




root = Tk()
root.title("University Prediction")
root.geometry("700x700")

root.minsize(600, 700)
root.maxsize(1080, 820)

root.configure(bg="lightblue")

Label_1 = Label(root, text="Welcome To Career Point", font=("Arial", 20, "bold"))
Label_1.grid(row=0, column=0, columnspan=2, pady=10)

# Country
country_label = Label(root, text="Country:")
country_label.grid(row=1, column=0, sticky="E")

Country = df2['country'].unique()
selected_Country = StringVar()
Country_list = OptionMenu(root, selected_Country, *Country)
Country_list.configure(bg="#FFFDD0", fg="Black", font=("Arial", 16), width=20)
Country_list.grid(row=1, column=1, sticky="W")

# City
city_label = Label(root, text="City:")
city_label.grid(row=2, column=0, sticky="E")

City = df2['city'].unique()
selected_City = StringVar()
City_List = OptionMenu(root, selected_City, *City)
City_List.configure(bg="#FFFDD0", fg="Black", font=("Arial", 16), width=20)
City_List.grid(row=2, column=1, sticky="W")

# Region
region_label = Label(root, text="Region:")
region_label.grid(row=3, column=0, sticky="E")

Region = df2['region'].unique()
selected_Region = StringVar()
Region_List = OptionMenu(root, selected_Region, *Region)
Region_List.configure(bg="#FFFDD0", fg="Black", font=("Arial", 16), width=20)
Region_List.grid(row=3, column=1, sticky="W")

# University Type
university_type_label = Label(root, text="University Type:")
university_type_label.grid(row=4, column=0, sticky="E")

University_type = df2['type'].unique()
selected_University_type = StringVar()
University_type_List = OptionMenu(root, selected_University_type, *University_type)
University_type_List.configure(bg="#FFFDD0", fg="Black", font=("Arial", 16), width=20)
University_type_List.grid(row=4, column=1, sticky="W")

# University Size
uni_size_label = Label(root, text="University Size:")
uni_size_label.grid(row=5, column=0, sticky="E")

Uni_Size = df2['size'].unique()
selected_Uni_Size = StringVar()
Uni_Size_List = OptionMenu(root, selected_Uni_Size, *Uni_Size)
Uni_Size_List.configure(bg="#FFFDD0", fg="Black", font=("Arial", 16), width=20)
Uni_Size_List.grid(row=5, column=1, sticky="W")

# Sports
sports_label = Label(root, text="Sports:")
sports_label.grid(row=6, column=0, sticky="E")

Sports = df2['sports'].unique()
selected_Sports = StringVar()
Option_menu = OptionMenu(root, selected_Sports, *Sports)
Option_menu.configure(bg="#FFFDD0", fg="Black", font=("Arial", 16), width=20)
Option_menu.grid(row=6, column=1, sticky="W")

# Percentage
percentage_label = Label(root, text="Grade:")
percentage_label.grid(row=7, column=0, sticky="E")

Percentage = df2['percentage'].unique()
selected_Grade = StringVar()
selected_Grade_menu = OptionMenu(root, selected_Grade, *Percentage)
selected_Grade_menu.configure(bg="#FFFDD0", fg="Black", font=("Arial", 16), width=20)
selected_Grade_menu.grid(row=7, column=1, sticky="W")

# Music
music_label = Label(root, text="Music:")
music_label.grid(row=8, column=0, sticky="E")

Music = df2['Music'].unique()
selected_Music = StringVar()
selected_Music_menu = OptionMenu(root, selected_Music, *Music)
selected_Music_menu.configure(bg="#FFFDD0", fg="Black", font=("Arial", 16), width=20)
selected_Music_menu.grid(row=8, column=1, sticky="W")





def find_university_names(given_university):
    matching_universities = []
    for university, university_name in university_list:
        if university == given_university:
            matching_universities.append(university_name)
    return matching_universities

def pred_math():
    reverse_dict_1 = {v: k for k, v in category_dict_1.items()}
    category_value_1 = selected_Country.get()
    if category_value_1 in reverse_dict_1:
        factorized_code_1 = reverse_dict_1[category_value_1]
    else:
        messagebox.showerror("Error", "Invalid country selected")
        return

    reverse_dict_2 = {v: k for k, v in category_dict_2.items()}
    category_value_2 = selected_City.get()
    if category_value_2 in reverse_dict_2:
        factorized_code_2 = reverse_dict_2[category_value_2]
    else:
        messagebox.showerror("Error", "Invalid city selected")
        return

    reverse_dict_3 = {v: k for k, v in category_dict_3.items()}
    category_value_3 = selected_Region.get()
    if category_value_3 in reverse_dict_3:
        factorized_code_3 = reverse_dict_3[category_value_3]
    else:
        messagebox.showerror("Error", "Invalid region selected")
        return

    reverse_dict_4 = {v: k for k, v in category_dict_4.items()}
    category_value_4 = selected_University_type.get()
    if category_value_4 in reverse_dict_4:
        factorized_code_4 = reverse_dict_4[category_value_4]
    else:
        messagebox.showerror("Error", "Invalid university type selected")
        return

    reverse_dict_5 = {v: k for k, v in category_dict_5.items()}
    category_value_5 = selected_Uni_Size.get()
    if category_value_5 in reverse_dict_5:
        factorized_code_5 = reverse_dict_5[category_value_5]
    else:
        messagebox.showerror("Error", "Invalid university size selected")
        return

    reverse_dict_6 = {v: k for k, v in category_dict_6.items()}
    category_value_6 = selected_Sports.get()
    if category_value_6 in reverse_dict_6:
        factorized_code_6 = reverse_dict_6[category_value_6]
    else:
        messagebox.showerror("Error", "Invalid sports selected")
        return

    reverse_dict_7 = {v: k for k, v in category_dict_7.items()}
    category_value_7 = selected_Grade.get()
    if category_value_7 in reverse_dict_7:
        factorized_code_7 = reverse_dict_7[category_value_7]
    else:
        messagebox.showerror("Error", "Invalid percentage selected")
        return

    reverse_dict_8 = {v: k for k, v in category_dict_8.items()}
    category_value_8 = selected_Music.get()
    if category_value_8 in reverse_dict_8:
        factorized_code_8 = reverse_dict_8[category_value_8]
    else:
        messagebox.showerror("Error", "Invalid music selected")
        return

    model = joblib.load('model.pkl')
    prediction = model.predict([[factorized_code_1, factorized_code_2, factorized_code_3, factorized_code_4, factorized_code_5, factorized_code_6, factorized_code_7, factorized_code_8]])
    given_university = prediction



    matching_universities = find_university_names(given_university)

    if matching_universities:
        prediction_result.config(text="Predicted University Names:\n" + "\n".join(matching_universities))
    else:
        prediction_result.config(text="No matching university found.")




prediction_button = Button(root, text="Submit", width=20, height=2, font=("Arial", 16), command=pred_math)
prediction_button.grid(row=9, column=0, columnspan=2, pady=10)


prediction_result = Label(root, text="", font=("Arial", 16), pady=10)
prediction_result.grid(row=10, column=0, columnspan=2, pady=10, sticky="E")

root.mainloop()
```
