```python
df['country'] = df['country'].factorize()[0]
df['city'] = df['city'].factorize()[0]
df['type'] = df['type'].factorize()[0]
df['region'] = df['region'].factorize()[0]
df['size'] = df['size'].factorize()[0]
df['sports'] = df['sports'].factorize()[0]
df['percentage'] = df['percentage'].factorize()[0]
df['Music'] = df['Music'].factorize()[0]
df.head()
```

```python
University_Names = df.drop(['University_Names'] , axis =1 , inplace = True)
```

```python
# Encode the categorical features
le = LabelEncoder()
for col in df.select_dtypes(include="category").columns:
    df[col] = le.fit_transform(df[col])

# Create the decision tree classifier
clf = DecisionTreeClassifier()

# Train the model
clf.fit(df.drop("University", axis=1), df["University"])

# Make predictions
predictions = clf.predict(df.drop("University", axis=1))

# Evaluate the model
print(clf.score(df.drop("University", axis=1), df["University"]))
```
